```python
import numpy as np
import pandas as pd

# params
frequencies_hz = [5, 10, 15, 20, 25]  
sequence_length = 1000              
sampling_rate = 100               
samples_per_class = 100          

data = []
labels = []

for label, freq in enumerate(frequencies_hz):
    period = int(sampling_rate / freq)  
    for _ in range(samples_per_class):
        spike_train = np.zeros(sequence_length, dtype=int)
        spike_indices = np.arange(0, sequence_length, period)
        spike_train[spike_indices] = 1 

        data.append(spike_train)
        labels.append(label)

df = pd.DataFrame(data)
df['label'] = labels

df.to_csv("synthetic_periodic_spike_dataset.csv", index=False)

print(df.head())
```
